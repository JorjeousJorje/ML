```python
import numpy as np
import matplotlib.pyplot as plt

from numpy import ndarray
from sklearn.datasets import fetch_openml
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import QuantileTransformer, Normalizer
from sklearn.tree import DecisionTreeClassifier

from TrainUtils import conduct_training, whole_in_one
%load_ext autoreload
%autoreload 2
```

```python
X, y = fetch_openml('mnist_784', return_X_y=True, as_frame=False)
y = y.astype(np.int32)
```

```python
quant_trans_uniform = QuantileTransformer(output_distribution="uniform")
```

```python
clf = make_pipeline(quant_trans_uniform, DecisionTreeClassifier(max_depth=5))
whole_in_one(clf=clf, X=X, y=y, train_size=5000)
```
