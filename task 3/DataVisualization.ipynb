```python

import numpy as np
import matplotlib.pyplot as plt

from time import time
from numpy import ndarray
from sklearn.datasets import fetch_openml
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import QuantileTransformer, Normalizer
from sklearn.linear_model import LogisticRegression
from sklearn.manifold import TSNE

from TrainUtils import conduct_training, whole_in_one
from PlotUtils import plot_mnist_samples, plot_embedding
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import (StandardScaler, 
                                   MaxAbsScaler,
                                   QuantileTransformer,
                                   MinMaxScaler,
                                   Normalizer)
%load_ext autoreload
%autoreload 2
```

```python
X, y = fetch_openml('mnist_784', return_X_y=True, as_frame=False)
```

```python
np.random.seed(100)
subsample_idc = np.random.choice(X.shape[0], 10000, replace=False)
X = X[subsample_idc, :]
y = y[subsample_idc]
```

# Unscaled data

```python
n_components = 2
tsne = TSNE(n_components=n_components, random_state=0)
tsne_result = tsne.fit_transform(X)
```

```python
plot_embedding(tsne_result, y)
```

# StandardScaler on data

```python
X_scaled = StandardScaler().fit_transform(X)
n_components = 2
tsne = TSNE(n_components=n_components, random_state=0)
tsne_result = tsne.fit_transform(X_scaled)
```

```python
plot_embedding(tsne_result, y)
```

# Normalizer on data

```python
X_scaled = Normalizer().fit_transform(X)
n_components = 2
tsne = TSNE(n_components=n_components, random_state=0)
tsne_result = tsne.fit_transform(X_scaled)
```

```python
plot_embedding(tsne_result, y)
```

# QuantileTransformer (uniform) on data

```python
quant_trans_uniform = QuantileTransformer(output_distribution="uniform")
```

```python
X_scaled = quant_trans_uniform.fit_transform(X)
n_components = 2
tsne = TSNE(n_components=n_components, random_state=0)
tsne_result = tsne.fit_transform(X_scaled)
```

```python
plot_embedding(tsne_result, y)
```

# QuantileTransformer (normal) on data

```python
quant_trans_normal = QuantileTransformer(output_distribution="normal")
```

```python
X_scaled = quant_trans_normal.fit_transform(X)
n_components = 2
tsne = TSNE(n_components=n_components, random_state=0)
tsne_result = tsne.fit_transform(X_scaled)
```

```python
plot_embedding(tsne_result, y)
```

```python
X_scaled = MinMaxScaler().fit_transform(X)
n_components = 2
tsne = TSNE(n_components=n_components, random_state=0)
tsne_result = tsne.fit_transform(X_scaled)
```
