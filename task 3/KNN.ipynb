```python
import numpy as np
import matplotlib.pyplot as plt

from numpy import ndarray
from sklearn.datasets import fetch_openml
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import QuantileTransformer, Normalizer
from sklearn.neighbors import KNeighborsClassifier

from TrainUtils import conduct_training, whole_in_one
%load_ext autoreload
%autoreload 2
```

# Загрузка MNIST

```python
X, y = fetch_openml('mnist_784', return_X_y=True, as_frame=False)
```

```python
quant_trans_uniform = QuantileTransformer(output_distribution="uniform")
```

# Classical KNN

```python
clf = make_pipeline(quant_trans_uniform, KNeighborsClassifier(n_neighbors=5))
whole_in_one(clf=clf, X=X, y=y, train_size=5000)
```

```python
clf = make_pipeline(quant_trans_uniform, KNeighborsClassifier(n_neighbors=10))
whole_in_one(clf=clf, X=X, y=y, train_size=5000)
```

```python
clf = make_pipeline(quant_trans_uniform, KNeighborsClassifier(n_neighbors=10, weights=lambda x: 1 / (1 + x)))
whole_in_one(clf=clf, X=X, y=y, train_size=5000)
```

```python
clf = make_pipeline(quant_trans_uniform, KNeighborsClassifier(n_neighbors=5, weights=lambda x: 1 / (1 + x)))
whole_in_one(clf=clf, X=X, y=y, train_size=5000)
```

```python
clf = make_pipeline(quant_trans_uniform, KNeighborsClassifier(n_neighbors=15))
whole_in_one(clf=clf, X=X, y=y, train_size=5000)
```
